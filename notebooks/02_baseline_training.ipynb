```python
from ultralytics import YOLO
import os

print("--- LOCAL TRAINING ACTIVATED ---")

# Verify the yaml is right where we left it
yaml_path = '../data/dataset.yaml'
if not os.path.exists(yaml_path):
    print("ERROR: Cannot find dataset.yaml. Make sure you are running this from the notebooks folder!")
else:
    print(f"SUCCESS: Found dataset map at {yaml_path}")
    
    # 1. Initialize the YOLOv8 Nano model (the absolute fastest version for CPU)
    model = YOLO('yolov8n.pt') 

    # 2. Start the Proof of Concept run!
    print("\nStarting CPU Training (1 Epoch)...")
    results = model.train(
        data=yaml_path,     
        epochs=1,            # Just 1 round to prove the pipeline works safely
        imgsz=640,          
        batch=4,             # Small batch size so we don't crash your laptop RAM
        device='cpu',        # Explicitly telling it to use the processor
        name='local_cpu_test' 
    )
    print("\nSUCCESS! Local training test complete.")
```
