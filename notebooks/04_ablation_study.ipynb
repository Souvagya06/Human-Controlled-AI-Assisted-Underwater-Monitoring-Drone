```python
from ultralytics import YOLO

# Load your custom YOLO-DarkWater architecture from the YAML we just made
model = YOLO("models/yolo-darkwater.yaml")

# Train the modified model
# Note: We don't use pretrained weights here because we changed the architecture!
results = model.train(
    data="data/dataset.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="yolo_darkwater_full",
    optimizer="AdamW" # Usually better for custom attention modules
)
```
